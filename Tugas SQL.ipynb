```python
import pymysql
import os
import pandas as pd

# Open database connection
conn = pymysql.connect(host='127.0.0.1', port=3306, user='root', passwd='', db='classicmodels')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#1. list of offices sorted by country, state, city
run("SELECT * FROM offices ORDER BY country, state, city")
```
