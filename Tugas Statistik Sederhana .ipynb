# Program Statistik

Data berikut merupakan data jumlah produksi padi dan palawija dengan kode ("data_padipalawija") dan data upah buruh tani ("data_upahtani"), data-data tersebut diambil dari tahun 2008-2014. Disini dilakukan analisis sederhana mulai dari jenis padi dan palawija yang memiliki produksi yang paling tinggi dan rendah serta ditahun berapa prouksi memiliki jumlah produksi yang paling tinggi. Kemudian untuk data upah tani, dilakukan pengamatan hasil visualisasi untuk didapat berapa upah paling tinggi dan rendah baik secara nominal maupun riilnya, tahun berapa nominal dan riil yang tinggi dan paling rendah. Kemudian dicari hubungan, ketika produksi padi dan palawija tinggi apakah nominal dan riil upah dari petani juga akan mengalami kenaikan yang sama sebaliknyapun begitu ketika produksi rendah apakah berdampak pada upah nominal dan riil petani. 

```python
import pandas as pd
```

```python
data_padipalawija = pd.read_csv('data-padi-palawija-1969-2014.csv', delimiter = ',', encoding = "ISO-8859-1")
data_upahtani = pd.read_csv('upah-nominal-riil-buruh-tani-2008-2014.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_padipalawija.count()
```

```python
data_upahtani.count()
```

```python
data_padipalawija.columns
```

```python
s = data_padipalawija.nilai.sum()
mi = data_padipalawija.nilai.min()
ma = data_padipalawija.nilai.max()
me = data_padipalawija.nilai.mean()

print('Jumlah Total nilai produksi 2008-2014:', "{:.2f}".format(s))
print('Nilai terkecil produksi 2008-2014:', "{:.2f}".format(mi))
print('Nilai terbesar produksi 2008-2014:', "{:.2f}".format(ma))
print('Nilai Rata-rata produksi 2008-2014:', "{:.2f}".format(me))
```

```python
data_padipalawija.groupby(['jenis_tanaman','tahun']).sum()['nilai']
```

```python
data_padipalawija.groupby('jenis_tanaman').mean()['nilai']
```

```python
data_padipalawija.groupby('tahun').sum()['nilai']
```

```python
ax = data_padipalawija.groupby('tahun').sum()['nilai']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tahun", figsize=(16,16), legend=True, fontsize=12) #function plot, jenis grafik nya pie , figsize ukuran grafik
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_padipalawija.groupby('jenis_tanaman').sum()['nilai']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tahun", figsize=(16,16), legend=True, fontsize=12) #function plot, jenis grafik nya pie , figsize ukuran grafik
ax.set_xlabel("Jenis Tanaman", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

3. Hubungan keduanya 
Jika ditelisik hubungan antara keduanya dimana jika jumlah produksi meningkat atau besar apakah nominal riil yang didapatkan juga akan mengalami peningkatan ?
-jika dilihat data padi dan palawija tahun 2011 memiliki jumlah produksi yang tinggi sebesar 4.190584e+08 sehingga jika dianalogikan maka seharusnya upah riil yang didapat juga akan tinggi, namun dari data yang didapat riil tahun 2011 malah mengalami penurunan dari tahun sebelumnya yang berjumlah 356023.0 menjadi 346458.0. Artinya tidak ada pengaruh kenaikan upah petani jika terjadi lonjakan jumlah produksi di suatu musim panen.

-namun analogi ini sempat terjadi direntan tahun 2008-2009 dimana jumlah produksi direntan tahun tersebut mengalami kenaikan dari 1.939904e+08 menjadi 2.055129e+08, sementara kenaikan upah riil dari 250649.0 menjadi 365678.0. artinya ada hubungan linear antara kedua data tersebut pada rentan tahun 2008-2009

-Sedangkan untuk nominal hubungan keduanya tidak terlihat dimana jumlah produksi hasil pertanian mengalami fluktuatif, namun angka jumlah nominal di rentan tahun 2008-2014 terus mengalami kenaikan tidak berpengaruh terhadap hasil produksi yang tidak stabil.
