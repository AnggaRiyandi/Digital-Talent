```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#1. Report the account representative for each customer
run("""
SELECT
    customerName,
    CONCAT(firstName, lastName) AS 'account representative'
FROM
    customers c
        INNER JOIN
    employees e ON c.salesRepEmployeeNumber = e.employeeNumber
""")
```

```python
#2.Report total payments for Atelier graphique.
run ("""SELECT
customers.customerNumber,
customers.customerName,
Sum(payments.amount)
FROM
customers
INNER JOIN payments ON payments.customerNumber = customers.customerNumber
WHERE
customers.customerNumber = 103
""")
```

```python
#3. Report the total payments by date
run ("""
SELECT
payments.paymentDate AS `Tanggal Pembayaran`,
Sum(payments.amount) AS Jumlah
FROM
payments
GROUP BY
payments.paymentDate
""")
```

```python
#4. Report the products that have not been sold

run("SELECT * FROM products p LEFT JOIN orderdetails o USING (productCode) WHERE orderNumber IS NULL")

```

```python
#5. List the amount paid by each customer.
run ("""SELECT
customers.customerName,
Sum(payments.amount) AS Jumlah
FROM
customers
INNER JOIN payments ON payments.customerNumber = customers.customerNumber
GROUP BY
customers.customerName
""")
```

```python
#6. How many orders have been placed by Herkku Gifts?
run ("""
SELECT
customers.customerName,
Count(orders.shippedDate) AS `Jumlah Pengiriman`
FROM
customers
INNER JOIN orders ON orders.customerNumber = customers.customerNumber
WHERE
customers.customerNumber = 167
GROUP BY
customers.customerName
""")
```

```python
#7. Who are the employees in Boston?

run ("""
SELECT 
    tabel1.employeeNumber, CONCAT(tabel1.firstName,' ',tabel1.lastName) employeeName, tabel2.city 
FROM
    employees tabel1
INNER JOIN
    offices tabel2
USING 
    (officeCode)
WHERE
    tabel2.city = 'Boston'
    """)
```

```python
#8.Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears   first

run ("""SELECT
customers.customerName,
Sum(payments.amount) AS Jumlah_Total_Pembayaran
FROM
customers
INNER JOIN payments ON payments.customerNumber = customers.customerNumber
WHERE
payments.amount > 100000
GROUP BY
customers.customerName
ORDER BY
Sum(payments.amount) DESC
""")
```

```python
#9. List the value of 'On Hold' orders.

run ("""
SELECT
orders.orderNumber,
orderdetails.productCode,
orders.orderDate,
orders.`status`,
orderdetails.priceEach
FROM
orders
INNER JOIN orderdetails ON orderdetails.orderNumber = orders.orderNumber
WHERE
orders.`status` = 'On Hold'
""")
```

```python
#10. Report the number of orders 'On Hold' for each customer.

run ("""
SELECT
customers.customerNumber,
customers.customerName,
Count(orders.`status`)
FROM
orders
INNER JOIN customers ON orders.customerNumber = customers.customerNumber
WHERE
orders.`status` = 'On Hold'
GROUP BY
customers.customerNumber,
customers.customerName
""")
```

# Many to Many

```python
#1. List products sold by order date.

run("""
SELECT
products.productCode AS `Kode Produk`,
products.productName AS `Nama Produk`,
orderdetails.quantityOrdered AS `Jumlah Pemesanan`,
orders.orderDate AS `Tanggal Pemesanan`
FROM
products
INNER JOIN orderdetails ON orderdetails.productCode = products.productCode
INNER JOIN orders ON orderdetails.orderNumber = orders.orderNumber


""")
```

```python
#2.List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

run ("""SELECT
orders.orderDate AS `Tanggal Pemesanan`,
orders.orderNumber AS `Kode Pemesanan`,
products.productName AS `Nama Produk`
FROM
products
INNER JOIN orderdetails ON orderdetails.productCode = products.productCode
INNER JOIN orders ON orderdetails.orderNumber = orders.orderNumber
WHERE
products.productName = '1940 Ford Pickup Truck'
ORDER BY
`Tanggal Pemesanan` DESC
""")
```

```python
# 3.List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

run("""
SELECT
orders.orderNumber AS `Kode Pemesanan`,
customers.customerName AS `Nama Pemesan / Customer`,
Sum((orderdetails.quantityOrdered * orderdetails.priceEach)) AS `value`
FROM
customers
INNER JOIN orders ON orders.customerNumber = customers.customerNumber
INNER JOIN orderdetails ON orderdetails.orderNumber = orders.orderNumber
GROUP BY
customers.customerName,
orders.orderNumber
HAVING
value > 25000
ORDER BY
`Kode Pemesanan` ASC

""")
```

```python
#4.Are there any products that appear on all orders?

run("""
SELECT 
    productName,
    COUNT(productCode) AS JML
FROM 
    orderdetails 
        INNER JOIN
    products USING(productCode)
GROUP BY 
    productCode 
HAVING 
    JML = (SELECT COUNT(*) FROM orders)
""")
```

```python
#5. List the names of products sold at less than 80% of the MSRP.

run ("""
SELECT
products.productName,
products.MSRP,
(products.MSRP*0.8),
orders.orderDate,
orderdetails.priceEach
FROM
products
INNER JOIN orderdetails ON orderdetails.productCode = products.productCode
INNER JOIN orders ON orderdetails.orderNumber = orders.orderNumber
WHERE
orderdetails.priceEach < 0.8*MSRP
GROUP BY
products.productCode
ORDER BY
(products.MSRP*0.8) DESC

""")
```

```python
#6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

run ("""
SELECT
products.productName,
products.buyPrice,
orders.orderDate,
orderdetails.priceEach,
(((priceEach - buyPrice) / buyPrice) * 100) AS markupPercentage
FROM
products
INNER JOIN orderdetails ON orderdetails.productCode = products.productCode
INNER JOIN orders ON orderdetails.orderNumber = orders.orderNumber
WHERE
orderdetails.priceEach >= 2*buyPrice
GROUP BY
products.productCode




""")
```

```python
#7.List the products ordered on a Monday.

run("""
SELECT 
    o.orderNumber,
    p.productCode As Kode_Produk,
    orderDate As Tanggal_Pesan_Senin,
    productName
FROM
    ((orderdetails
        INNER JOIN
     products p USING (productCode))
        INNER JOIN
    orders o USING (orderNumber))
WHERE
    DAYNAME(orderDate) = 'Monday'
ORDER BY
    orderNumber
""")
```

```python
#8.What is the quantity on hand for products listed on 'On Hold' orders?

run ("""
SELECT
products.productName,
orders.orderNumber,
orderdetails.productCode,
products.quantityInStock,
orders.`status`
FROM
orders
INNER JOIN orderdetails ON orderdetails.orderNumber = orders.orderNumber
INNER JOIN products ON orderdetails.productCode = products.productCode
WHERE
orders.`status` = 'On Hold'
""")
```
